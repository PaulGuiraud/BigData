**Après avoir listé chaque couple de code en fonction du code état associé, nous enlevons les couples qui comportent un 'nan'.**  
**Le but étant de pouvoir avoir une source d'ancien code facilement remplaçables.**

```python
code_rpl_etat = [(a,b,c) for a,b,c in list_code_rpl_etat if (a!= 'nan') and (b != 'nan')]
#code_rpl_etat
```

```python
unique_trouples_etat = list(set(code_rpl_etat))
print("Nous avons donc ",len(unique_trouples_etat)," différents trouples")
```

**Nous avons donc ici, pour chaque nouveau code, la liste des couples (Ancien code, Code état).**

```python
# dict new to old 
unique_new_etat = np.unique([b for a,b,c in unique_trouples_etat])
by_new_etat = [('New code :',new_code,[('Old codes :',old_code,'N. Etat :',nb_etat) for old_code,b,nb_etat in unique_trouples_etat if b == new_code]) for new_code in unique_new_etat]
## vraie liste avec les nouveaux codes états et chaque couple (Ancien code / code état) qui lui sont associés. 
by_new_etat
```

**Nous créons ensuite un dictionnaire afin de pouvoir séparer les anciens codes qui ne menent qu'à un nouveau code quelque soit le code état de ceux qui peuvent avoir été séparés.**

```python
## Lors du dictionnaire de dictionnaire, si un nouveau code est l'arrivée d'un ancien code sous plusieurs code état, celui apparaitra 1seule fois 
## on veut juste récuperer la liste des anciens codes ne pointant que vers 1 nouveau code. 
by_new_etat_dict = dict([(new_code,{old_code : nb_etat for old_code,b,nb_etat in unique_trouples_etat if b == new_code}) for new_code in unique_new_etat])
by_new_etat_dict
```

```python
unique_old_etat = np.unique([a for a,b,c in unique_trouples_etat])
```

```python
dict_old_code_remplacable_c_to_v = {}
dict_old_code_remplacable_v_to_c = {True : [] ,False : []}
for old_code in unique_old_etat :
    result = 0 
    for new_code in by_new_etat_dict.keys():
        if old_code in by_new_etat_dict[new_code]:
            result += 1
    dict_old_code_remplacable_v_to_c[result == 1].append(old_code)
    dict_old_code_remplacable_c_to_v[old_code] = (result == 1)
```

**Nous avons donc 590 anciens codes ne posant pas problème, nous allons pouvoir les utiliser. Pour cela nous créerons 3 dictionnaires :**  
- **Le premier sera les anciens codes remplaçables associés à leurs nouveaux ccodes**
- **Le second contiendra ceux qui ne sont pas remplçables, associés à eux même.**
- **Le troisième contiendra les codes non utilisés (ceux dont les lignes contenait des 'nan') (on se rendra compte qu'il n'y en a pas.)**

```python
good_replacment = dict([(old_code,list(np.unique([new_code for a,new_code,nb_etat in unique_trouples_etat if a == old_code]))) for old_code in dict_old_code_remplacable_v_to_c[True]])
```

```python
good_replacment ## un dico d'anciens codes vers les nouveaux dont on est sur qu'on peut les remplacer 
```

```python
bad_replacment = dict([(a,[a]) for a in dict_old_code_remplacable_v_to_c[False]])
```
